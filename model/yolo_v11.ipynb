```python
model1 = YOLO('model_test.pt')
model2 = YOLO('runs/detect/train32/weights/best.pt')
```

```python
classes1 = list(model1.names.values())
classes2 = list(model2.names.values())
```

```python
combined_classes = classes1 + [cls for cls in classes2 if cls not in classes1]

# Initialize a new YOLOv8 model with the architecture of one of the pretrained models
combined_model = YOLO('model_test.pt')
combined_model.model.nc = len(combined_classes)
combined_model.model.names = {i: name for i, name in enumerate(combined_classes)}
```

```python
weights1 = model1.model.state_dict()
weights2 = model2.model.state_dict()
combined_weights = combined_model.model.state_dict()
```

```python
for key in weights1:
    if key in combined_weights:
        combined_weights[key] = weights1[key]
for key in weights2:
    if key in combined_weights:
        combined_weights[key] = weights2[key]
```

```python
combined_model.model.load_state_dict(combined_weights, strict=False)

# Save the combined model
torch.save(combined_model.model.state_dict(), 'yolov11_combined_model.pt')

# Perform inference with the combined model
#results = combined_model(source="video.mp4", show=True, conf=0.1, save=True)
```

```python

def predict_image(yolo_model, img):

    results = yolo_model.predict(img, conf=0.4)  # Run inference on the image

    detections = results[0].boxes  # Get predictions
    bboxes = []  # List to store bounding boxes
    bbox_metrics = []  # List to store metrics

    # Проход по всем предсказаниям
    for det in detections:
        # Unpack the detection
        x1, y1, x2, y2 = det.xyxy[0]  # Bounding box coordinates
        conf = det.conf[0]  # Confidence scores
        cls = det.cls[0]  # Class index
    
        bboxes.append([x1.item(), y1.item(), x2.item(), y2.item()])  # Store bounding box coordinates
        bbox_metrics.append({'confidence': conf.item(), 'class': int(cls.item())})  # Store metrics

    # Draw bounding box on the image
    #img = cv2.imread(img)  # Read the original image
    #cv2.rectangle(img, (int(x1), int(y1)), (int(x2), int

    return bboxes, bbox_metrics


```

```python
import cv2
import numpy as np

# Extract the mask from predictions
mask = pred[0].masks.data[0]  # Assuming you want the first mask
mask_image = Image.fromarray(mask.byte().cpu().numpy() * 255)  # Convert mask to binary image

# Convert the mask to a NumPy array and find contours
mask_np = np.array(mask_image)  # Convert to NumPy array
contours, _ = cv2.findContours(mask_np, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Create a blank mask for the cutout
cutout_mask = np.zeros_like(mask_np, dtype=np.uint8)  # Ensure the mask is of type uint8

# Draw the contours on the blank mask
cv2.drawContours(cutout_mask, contours, -1, (255), thickness=cv2.FILLED)

# Load the original image
original_image = Image.open(img)
original_image_np = np.array(original_image)  # Convert to NumPy array

# Ensure the original image is of type uint8
if original_image_np.dtype != np.uint8:
    original_image_np = original_image_np.astype(np.uint8)

# Resize the mask to match the original image size
cutout_mask = cv2.resize(cutout_mask, (original_image_np.shape[1], original_image_np.shape[0]), interpolation=cv2.INTER_NEAREST)

# Create a cutout image using the mask
cutout_image = cv2.bitwise_and(original_image_np, original_image_np, mask=cutout_mask)

# Convert the cutout image back to PIL format for display
cutout_image_pil = Image.fromarray(cutout_image)

# Display the cutout image
plt.imshow(cutout_image_pil)
plt.axis('off')  # Hide axes
plt.show()
```
